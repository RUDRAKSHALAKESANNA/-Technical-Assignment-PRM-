```python
import pandas as pd 4
```

```python
order_reports=pd.read_excel("/content/Company X - Order Report.xlsx")
order_reports
```

```python
SKU_master=pd.read_excel("/content/Company X - SKU Master.xlsx")
SKU_master
```

```python
pincode_mapping=pd.read_excel("/content/Company X - Pincode Zones.xlsx")
pincode_mapping
```

```python
common_column="SKU"
merged_df = pd.merge(SKU_master, order_reports, on=common_column)

merged_df
```

```python
merged_df.rename(columns={'ExternOrderNo': 'Order ID'}, inplace=True)
merged_df
```

```python
invoice=pd.read_excel("/content/Courier Company - Invoice.xlsx")
invoice
```

```python
common_column="Order ID"
merged_df1 = pd.merge(invoice, merged_df, on=common_column)

merged_df1
```

```python
common_column="Warehouse Pincode"
merged_df2 = pd.merge(pincode_mapping, merged_df1, on=common_column)

merged_df2
```

```python
merged_df2['Total Weight'] = merged_df2['Order Qty'] * merged_df2['Weight (g)']
merged_df2
```

```python
Rates=pd.read_excel("/content/Courier Company - Rates.xlsx")
Rates
```

```python
def get_delivery_zone(row):
    if row['Zone_x'] =="a":
        return 'A'
    elif row['Zone_x']=="b":
        return 'B'
    elif row['Zone_x']=="c":
        return 'C'
    elif row['Zone_x']=="d":
        return 'D'
    elif row['Zone_x']=="e":
        return 'E'

merged_df2['Delivery Zone'] = merged_df2.apply(get_delivery_zone, axis=1)
merged_df2
```

```python
merged_df2.rename(columns={'Delivery Zone': 'Zone'}, inplace=True)
merged_df2
```

```python
common_column="Zone"
merged_df3 = pd.merge(merged_df2, Rates, on=common_column)

merged_df3
```

```python
def calculate_cod_charge(row):
    if row['Payment Mode'] == 'Prepaid':
        return 0
    else:
        if row['Item Price(Per Qty.)'] <= 300:
            return 15
        else:
            return 0.05 * row['Item Price(Per Qty.)']


merged_df3['COD Charge'] = merged_df3.apply(calculate_cod_charge, axis=1)
merged_df3
```

```python
merged_df3["courier forwad charges"]=merged_df3["Item Price(Per Qty.)"]*merged_df3["Weight Slabs"]+merged_df3["Forward Fixed Charge"]
merged_df3
```

```python
merged_df3["courier rto cahrges"]=merged_df3["Item Price(Per Qty.)"]*merged_df3["Weight Slabs"]+merged_df3["RTO Fixed Charge"]
merged_df3
```

```python
merged_df3
```

```python
merged_df3[" courier total charge"]=merged_df3["COD Charge"]+merged_df3["courier forwad charges"]+merged_df3["courier rto cahrges"]
merged_df3
```

```python
merged_df3["X forwad charges"]=merged_df3["Item Price(Per Qty.)"]*merged_df3["Weight Slabs"]+merged_df3["Forward Additional Weight Slab Charge"]
merged_df3
```

```python
merged_df3["X rto cahrges"]=merged_df3["Item Price(Per Qty.)"]*merged_df3["Weight Slabs"]+merged_df3["RTO Additional Weight Slab Charge"]
merged_df3
```

```python
merged_df3[" X total charge"]=merged_df3["COD Charge"]+merged_df3["X forwad charges"]+merged_df3["X rto cahrges"]
merged_df3
```

```python
merged_df3.to_csv('merged_df3.csv')
```
